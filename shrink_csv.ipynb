<h3> Load libraries and dataset .csv file</h3>

```python
import pandas as pd
import numpy as np

#will need to update address of UNCC_sampled.csv to local address
data = pd.read_csv("../Visual_Analytics_Final_Project/UNCC_sampled.csv")
```

<h3>The dataset</h3>

```python
#displays data
data
```

<h3>Dataset size and list of named columns</h3>

```python
print('(rows, columns) ', data.shape)
data.columns
```

<h3>Code to extract specific columns into a new .csv file</h3>

```python
#update this list with the column names of the data paramater to be used
columns_to_use = ['Date','GeoSegmentation States']
#new dataframe of just the sesired columns
small_data = data[columns_to_use]
#update output file name as needed
small_data.to_csv('small_UNCC.csv', sep=',')
#prints new size of data
print(small_data.shape)
```

<h3>Drops rows from data set that are empty in a specific set of columns</h3>

```python
#drops rows that have Nan values(no value) in any column in the subset
cleaned_small_data = small_data.dropna(subset=['GeoSegmentation States'])
#prints the new size of the data
print(cleaned_small_data.shape)
#update output file name as needed
cleaned_small_data.to_csv('cleaned_small_UNCC.csv', sep=',')
```

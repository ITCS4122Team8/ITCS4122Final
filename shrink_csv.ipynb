```python
#pull subset
columns_to_use = ['Visit Number',
          'Form Engagement 1 (ev11) (event11)',
          'Form Engagement 2 (ev12) (event12)',
          'GeoSegmentation States',
            'Page Views','Week_Day']

df = data[columns_to_use]

#rename columns
df = df.rename(index=str, columns={"Visit Number": "Visit_Number",
                              "Form Engagement 1 (ev11) (event11)": "Form_Engagement_1",
                              "Form Engagement 2 (ev12) (event12)": "Form_Engagement_2",
                              "GeoSegmentation States": "state",
                                  'Page Views': 'Page Views'})


#fill in blank values
df['state'].fillna("no state", inplace=True)
df['Form_Engagement_1'].fillna(0, inplace=True)
df['Form_Engagement_2'].fillna(0, inplace=True)
df['Visit_Number'].fillna(0, inplace=True)

#removes rows where users did not engage forms
df = df[df.Form_Engagement_1 != 0]
df = df[df.Form_Engagement_1 != 2]
df = df[df.Form_Engagement_2 != 0]
df = df[df.Form_Engagement_2 != 2]

#removes places not in the US
df = df[df.state.str.contains("(united states)")]

#sorts df by visit number
df.sort_values(by=['Visit_Number'], inplace= True)

#replaces " (united states)" with ''
df['state'] = df['state'].str.replace(" \(united states\)",'')

#groups by state and counts for each
df = df.groupby('state').sum()

print(df.shape)
df
```
